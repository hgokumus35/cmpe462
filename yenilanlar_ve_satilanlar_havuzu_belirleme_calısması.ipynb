```python
#kıyaslıyıcağımız iki günü burada önceki ve sonraki olmak üzere pandasın dataframe yapısını kullanıyoruz çünkü gruplandırmalar
# ve filitrelendirmeler daha rahat yapılıyor
import pandas as pd

path_onceki='/Users/Efe can DEMİR/Desktop/cars/ilanlar/June_16_21__2235.json'
onceki=pd.read_json(path_onceki)

path_sonraki='/Users/Efe can DEMİR/Desktop/cars/ilanlar/June_18_21__2227.json'
sonraki=pd.read_json(path_sonraki)
```

```python
sonraki
```

```python
#her ilanın kendine özgü ilan no olduğu için ilan nolarını index numarası yaptık
onceki.set_index('id',inplace=True)
sonraki=sonraki.set_index('id')
sonraki
```

```python
sonraki.loc[~sonraki.index.isin(onceki.index)]
#sonraki içinde olup önceki içinde olmayan ilan numaralarını buradan buluyoruz bunlar neyi konan ilanlarımız olucak
```

```python
# bu kadar yeni ilan ? oncekinde olmayan sonrakindeki her ilan yeni mi? hayır çünkü bazen eski ilan sahipleri ilanlarını çekebilirler
#Yada ilanın süresi bitebilir. Bu tür ilanlar bizim için sorun yaratır onları değerlendirmeye almamamız lazım
```

```python
yeni=sonraki.loc[sonraki.index>max(onceki.index)]
yeni
#Bu yüzden ilan tanımımızı önceki dosyamızdaki en büyük ilan numrasından daha büyük olan ilanlara yeni ilanlar diyelim
```

```python
#Bu ilanlar arasindan ilan numarasi maximumdan büyük olmayanlari karalisteye aliyoruz 
#çünkü bunlar kaldırılıp yeniden koyulan ilanlar
#Satılanlar havuzumu bu ilanlardan arındıryoruz
karaliste=sonraki.loc[~sonraki.index.isin(onceki.index)]
karaliste=karaliste.loc[karaliste.index<=max(onceki.index)]
karaliste
```

```python
#Satılanlar önceki içinde olup sonraki içinde olmayanlar olarak değerlendiriyoruz daha arındırma yapmadık
satilan=onceki.loc[~onceki.index.isin(sonraki.index)]
satilan
#Arındırma için ilana giriş tarihleri addate olarak var çıkış tarihlerinide dosya adı olarak çekerken kayıt ettim
```

```python
#Bu kısımda tarih bilgisini dosyanın adından alıyorum
from datetime import datetime

satilma_tarihi=datetime.strptime(path_sonraki[42:-11], '%B_%d_%y')
satilma_tarihi

```

```python
#Bu ikisi arasından ne kadar zaman geçti diye bakıyorum ve bu süreyi kaç günde değişkeni 
#adında yeni bir colum olarak dataframe'e ekliyoruz
satilan['kac_gunde']=satilan.adDate.apply(lambda x: (satilma_tarihi-datetime.strptime(x,'%Y-%m-%d')).days) 
satilan
```

```python
satilan7=satilan[satilan.kac_gunde<=7]
satilan7
```
