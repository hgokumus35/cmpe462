```python
import pandas as pd
from datetime import datetime
from glob import glob

#convert axis true olursa (varsayilan), satirlar sutunlar yer degistirir
satilanlar_havuzu=pd.read_json('/Users/Efe can DEMİR/Desktop/cars/satilanlar_havuzu.json',orient='index',convert_axes=False)
satilanlar_havuzu
```

```python
satilanlar_havuzu.index=pd.to_numeric(satilanlar_havuzu.index,errors='coerce') #coerce'le numerik yapmaya zorla
satilanlar_havuzu=satilanlar_havuzu.dropna() #sayi yapamazsan dropla
```

```python
dosya_listesi=glob("/Users/Efe can DEMİR/Desktop/cars/ilanlar/*.json")
dosya_listesi.sort(key=lambda x: datetime.strptime(x[42:-5], '%B_%d_%y__%H%M'))

path_onceki=dosya_listesi[-2] 
path_sonraki=dosya_listesi[-1]                            
      
onceki=pd.read_json(path_onceki)
sonraki=pd.read_json(path_sonraki)
onceki=onceki.set_index('id')
sonraki=sonraki.set_index('id')
    
yeni=pd.DataFrame() 
yeni=sonraki.loc[sonraki.index>max(onceki.index)]

karaliste=pd.DataFrame()   
karaliste=sonraki.loc[~sonraki.index.isin(onceki.index)]
karaliste=karaliste.loc[karaliste.index<=max(onceki.index)]
    
satilan=pd.DataFrame()
satilan=onceki.loc[~onceki.index.isin(sonraki.index)]

satilma_tarihi=datetime.strptime(path_sonraki[42:-11], '%B_%d_%y')
    
satilan['kac_gunde']=satilan.adDate.apply(lambda x: (satilma_tarihi-datetime.strptime(x,'%Y-%m-%d')).days) 

satilan7=satilan[satilan.kac_gunde<=7]
satilanlar_havuzu=pd.concat([satilanlar_havuzu,satilan7])
satilanlar_havuzu=satilanlar_havuzu.loc[~satilanlar_havuzu.index.isin(karaliste.index)] 

satilanlar_havuzu=satilanlar_havuzu[~satilanlar_havuzu.index.duplicated(keep='first')]
#tekrar ilanlar varsa bu ilanlardan kurtulma için son satırı kullanıyoruz


```

```python
yeni['id']=yeni.index
sonuclar = pd.merge(yeni, arac_istatistikleri, how='left',on=['brand','model', 'notes', 'proDate'])
sonuclar=sonuclar.set_index('id')
sonuclar
```

```python
sonuclar_v2=sonuclar[['brand','model','notes','proDate',('price', 'count'),('kac_gunde', 'mean'),
                               'price',('price', 'mean'), ('price', '50%'), #yuzde elli medyan degerini aldik
                               'km',('km', 'mean'),('km', '50%')]]
sonuclar_v2
```

```python
#bazı araçlar istediğiz aralıkta satılmamış olabilirler
#bu araçların değerleri NaN geliyor onları görmek ve değerlendirmek istemiyoruz
#Bu yüzden onları çıkartıyoruz
sonuclar_v3=sonuclar_v2.dropna()
#NaNlar çıktıktan sonra v2 de olup v3te olmayanlarıda aztelep görenler olarak ayırıyorum

aztalep_araclar = sonuclar_v2[~sonuclar_v2.index.isin(sonuclar_v3.index)]
```

```python
#sutun isimlerini turkclestiriyoruz
#sonuclar[['brand','model','notes','proDate',('price', 'count'),('kac_gunde', 'mean'),
#                               'price',('price', 'mean'), ('price', '50%'),
#                               'km',('km', 'mean'),('km', '50%')]]


sonuclar_v3.columns=['marka','model','detay','yil','sayi','gun',
                    'FIYAT','ortF','F50',
                    'KM','ortKM','KM50']

sonuclar_v3
```

```python
sonuclar_v3.sayi=sonuclar_v3.sayi.apply(lambda x:int(x))
sonuclar_v3.gun=sonuclar_v3.gun.apply(lambda x:round(x,1))
sonuclar_v3.ortF=sonuclar_v3.ortF.apply(lambda x:int(x))
sonuclar_v3.F50=sonuclar_v3.F50.apply(lambda x:int(x))
sonuclar_v3.ortKM=sonuclar_v3.ortKM.apply(lambda x:int(x))
sonuclar_v3.KM50=sonuclar_v3.KM50.apply(lambda x:int(x))
```

```python
#araçların fiyatlarını ve kmlerini bulduğumuz ortadeğerle karşılaştırıcaz
#buda bize fırsat ve km değerini vericek
sonuclar_v3['FIRSAT']=(sonuclar_v3.F50-sonuclar_v3.FIYAT)/sonuclar_v3.F50
sonuclar_v3.FIRSAT=sonuclar_v3.FIRSAT.apply(lambda x:int(x*100))
sonuclar_v3['KM_DEGERI']=(sonuclar_v3.KM50-sonuclar_v3.KM) #MEDYAN KM ILE ARACIN KMS ARASI FARK
sonuclar_v3
```

```python
sonuclar_v4=sonuclar_v3.sort_values(by='FIRSAT',ascending=False) #ascending varsayilan olarak true geliyor
sonuclar_v4
```

```python
sonuclar_v5=sonuclar_v4[(sonuclar_v4['FIYAT'] >=0.70*sonuclar_v4['F50'] ) & (sonuclar_v4['FIYAT'] <=1.30*sonuclar_v4['F50'] )]
sonuclar_v5
#hatali araclari degerlendirmeden çikarmak için ortlamanın %70 ile %130' arasındaki araçları aldık
#çünkü yanlış yazılmalar ve çok pahalı araçları görmek istemedik
```
